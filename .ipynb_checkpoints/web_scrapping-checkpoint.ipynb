```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### WEB SCRAPPING

```python
url= "https://www.worldometers.info/gdp/gdp-by-country/"
```

```python
headers={'User-Agent': 'Mozilla/5.0'}
```

```python
response=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
table=soup.find_all("table")
```

### Load data into dataframe

```python
data=pd.read_html(str(table))
```

```python
df=data[0]
```

```python
df.info()
```

```python
df
```

```python
df.drop(columns='#',inplace=True)
```

```python
df
```

### Preprocess the Data

```python
df["Share of World GDP"]=df["Share of World GDP"].str.replace("%","").str.strip()
```

```python
df["GDP per capita"]=df["GDP per capita"].str.replace("$","").str.replace(",","").str.strip()
```

```python
df["Share of World GDP per"]=df["Share of World GDP"]
```

```python
df.drop(["Share of World GDP"],axis=1,inplace=True)
```

```python
df["GDP Growth"]=df["GDP Growth"].str.replace("%","").str.strip()
```

```python
df["GDP (nominal, 2023)"]=df["GDP (nominal, 2023)"].str.replace("$","").str.strip()
```

```python
df["GDP (nominal, 2023)"]=df["GDP (nominal, 2023)"].str.replace(",","").str.strip()
```

```python
df["GDP (abbrev.)"]=df["GDP (abbrev.)"].str.replace("$","").str.strip()
```

```python
df.info()
```

```python
float(df["GDP (abbrev.)"][1].replace("trillion","").strip())
```

```python
a=df["GDP (abbrev.)"][1]
a
```

```python
convert_gdp=[]
for i in df["GDP (abbrev.)"]:
    if "trillion" in i:
        i=i.replace("trillion","").strip()
        i =float(i)*100000000000
    elif "billion" in i:
        i=i.replace("billion","").strip()
        i=float(i)*1000000000
    elif "million" in i:
        i=i.replace("million","").strip()
        i=float(i)*1000000
    convert_gdp.append(i)

df["GDP (abbrev.)"]=convert_gdp
    
```

```python
a='-0.27'
a=a.replace("-","").strip()
a
```

```python
convert=[]
for i in df["GDP Growth"]:
    if "−" in i:
        i=i.replace("−","").strip()
        i=float(i)*-1
    else:
        i=float(i)*1
    convert.append(i)
df["GDP Growth"]=convert
        
```

```python
df["GDP (nominal, 2023)"]=df["GDP (nominal, 2023)"].astype(float)
df["GDP Growth"]=df["GDP Growth"].astype(float)
df["Share of World GDP per"]=df["Share of World GDP per"].astype(float)
df["GDP per capita"]=df["GDP per capita"].astype(float)
```

### Clustering countries based on gdp and growth

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data=df[["Country","GDP (nominal, 2023)","GDP Growth"]]
```

```python
data
```

##### Features on which we cluster

```python
X=data[["GDP (nominal, 2023)","GDP Growth"]]
```

```python
X
```

```python
scaler=StandardScaler()
```

```python
x_new=scaler.fit_transform(X)
```

```python
x_new
```

```python
kmeans=KMeans(n_clusters=4,random_state=42,n_init=10)
```

```python
cluster=kmeans.fit_predict(x_new)
```

```python
data["cluster"]=cluster
```

### Outcome of Clustering

```python
data
```

#### Stable Economy Countries

```python
data[data["cluster"]==2]
```

#### Growing Economy Countries

```python
data[data["cluster"]==1]
```

#### Struggling Economy Countries

```python
data[data["cluster"]==0]
```

#### Very Fast Growing Countries

```python
data[data["cluster"]==3]
```
